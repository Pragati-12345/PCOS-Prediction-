```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score, classification_report
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.impute import SimpleImputer
```

```python
data = pd.read_csv(r"C:\Users\Dell\data_cleaned_results.csv")
data.head()
```

```python
del data['City']
del data['relocated city']
del data['Unnamed: 0']
```

```python

data.head()
```

```python

data['PCOS_label'] = None
data = data.set_index('PCOS_label')
data = data.reset_index()
data

```

```python
def label(row):
    if row['PCOS'] == 'Yes':
        return 1
    else:
        return 0

data['PCOS_label'] = data.apply(lambda row: label(row), axis=1)

data.head()
```

```python

PCOS_check = dict(zip(data.PCOS_label.unique(), data.PCOS.unique()))
PCOS_check
```

```python

x = data.drop(['PCOS_label', 'PCOS'], axis=1)
y = data.PCOS_label
```

```python
#convert non numeric rows to 0/1
x = pd.get_dummies(x, drop_first=True)
#checks for row left unconverted
print(x.dtypes)

```

```python
xtrain, xtest, ytrain, ytest = train_test_split(
    x, y, test_size=0.25, random_state=0,stratify=y)

```

```python
print("Original:", y.value_counts())
print("Train:", ytrain.value_counts())
print("Test:", ytest.value_counts())
```

```python
imputer = SimpleImputer(strategy='mean')
xtrain = imputer.fit_transform(xtrain)
xtest = imputer.transform(xtest)
```

```python
sc_x = StandardScaler()
xtrain = sc_x.fit_transform(xtrain)
xtest = sc_x.transform(xtest)
```

```python
print(xtrain[0:10, :])

classifier = LogisticRegression(random_state=0)

```

```python
classifier.fit(xtrain, ytrain)
y_pred = classifier.predict(xtest)
```

```python
cm = confusion_matrix(ytest, y_pred)

print("Confusion Matrix : \n", cm)

print("Accuracy : ", accuracy_score(ytest, y_pred))
```
